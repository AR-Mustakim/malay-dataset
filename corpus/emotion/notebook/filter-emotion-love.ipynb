```python
import json

with open('love-twitter-malaysia.json') as fopen:
    love = json.load(fopen)
len(love)
```

```python
negates = ['salah', 'naik kepala', 'benci', 'bodoh', 'sakai', 'shame', 'shaming', 'menangis',
          'berdarah', 'tak boleh', 'sedih', 'nangis', 'merasa cukup', 'porn', 'letgo', 'let go',
          'bilik sewa', 'rumah sewa', ':(', 'huhu', 'hadis']
```

```python
filtered = [j for j in love if not any([n in j.lower() for n in negates])]
```

```python
len(filtered)
```

```python
with open('love-twitter-malaysia.json', 'w') as fopen:
    json.dump(filtered, fopen)
```
