```python
import json

with open('sadness-twitter-malaysia.json') as fopen:
    sadness = json.load(fopen)
len(sadness)
```

```python
negates = ['haha', 'syok', 'happy']
```

```python
filtered = [j for j in sadness if not any([n in j.lower() for n in negates])]
```

```python
len(filtered)
```

```python
filtered
```

```python
with open('sadness-twitter-malaysia.json', 'w') as fopen:
    json.dump(filtered, fopen)
```
