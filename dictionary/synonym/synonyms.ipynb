```python
translators = translate_new.Translate_Concurrent(batch_size = 20, from_lang = 'en', to_lang = 'ms')
```

```python
with open('ppdb_synonyms.txt') as fopen:
    synonyms = list(filter(None, fopen.read().split('\n')))
    
synonyms = [w.strip().replace(' ', ' [EENNDD] ') for w in synonyms]
```

```python
t = translators.translate_batch(synonyms)
```

```python
import json

with open('synonyms-translated.json', 'w') as fopen:
    json.dump(t, fopen)
```

```python
bahasa = list(zip(*t))[1]
bahasa = set(bahasa)
bahasa = [i.split('[EENNDD]') for i in bahasa]
bahasa = [i for i in bahasa if '[DAN]' not in i and len(i) == 2]
bahasa = [i for i in bahasa if i[1] != '' and i[0] != '']
bahasa = [[i[0].strip(), i[1].strip()] for i in bahasa]
```
