```python
import json

with open('anger-twitter-malaysia.json') as fopen:
    anger = json.load(fopen)
len(anger)
```

```python
negates = ['haha']
```

```python
filtered = [j for j in anger if not any([n in j.lower() for n in negates])]
```

```python
len(filtered)
```

```python
filtered
```

```python
with open('anger-twitter-malaysia.json', 'w') as fopen:
    json.dump(filtered, fopen)
```
