```python
cookies = {'klk_currency': 'MYR'}
url = 'https://www.klook.com/xos_api/v1/usrcsrv/search/all_activities?country_id=19&template_id=1&start=%d&size=24'
```

```python
index = 1
results = []
while True:
    print(index)
    response = requests.get(url %(index), headers = combined, 
                            cookies = cookies, timeout = 60)
    result = response.json()
    if 'activities' not in result['result']:
        break
        
    results.extend(result['result']['activities'])
    index += 1
```

```python
len(results)
```

```python
import json

with open('malaysia-klook-list.json', 'w') as fopen:
    json.dump(results, fopen)
```

```python
for i in range(84, len(results)):
    result = results[i]
    link = 'https://www.klook.com/en-MY/activity/%d-%s/'%(result['id'], result['url_seo'])
    print(link)
    
    response = requests.get(link, 
                        headers = combined, cookies = cookies, timeout = 60)
    soup = BeautifulSoup(response.content, 'html.parser')
    all_scripts = soup.find_all('script')
    data = None

    for start, end, obj in jsonfinder(all_scripts[-2].text):
        if obj:
            if 'id' in obj:
                data = obj
    if data is None:
        print(i)
                
    results[i]['full-description'] = data
```
