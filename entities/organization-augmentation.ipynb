```python
import numpy as np
import json
```

```python
with open('malaysia-organization.json') as fopen:
    org = json.load(fopen)
```

```python
import random

def generate_org(length):
    if str(length - 1) not in org:
        o = ' '.join(random.sample(org['free'], length))
    else:
        o = random.choice(org[str(length - 1)])
    return o
```

```python
with open('entities-data-v4.json') as fopen:
    entities = json.load(fopen)
```

```python
results = []
i = 0
while i < len(entities['label']):
    r = []
    if entities['label'][i] == 'organization':
        while entities['label'][i] == 'organization':
            r.append(i)
            i += 1
        print(r)
        results.append(r)
    i += 1
```

```python
import math

def generate_index(l, name, texts, labels, length):
    cp, indices = [], []
    b = length - len(l)
    left = math.ceil(b / 2)
    right = b - left
    minus = l[0] - left
    if minus < 0:
        absolute = np.abs(minus)
        right += absolute
        left -= absolute
    for i in range(l[0] - left, l[0]):
        cp.append(texts[i])
        indices.append(labels[i])
    cp.extend(name)
    indices.extend([labels[i] for i in l])
    try:
        for i in range(l[-1] + 1, l[-1] + right + 1):
            cp.append(texts[i])
            indices.append(labels[i])
    except Exception as e:
        print(e)
        pass
    return cp, indices
```

```python
X, Y = [], []

for r in results:
    for _ in range(3):
        x, y = generate_index(r, generate_org(len(r)).split(), entities['text'], entities['label'], 20)
        X.append(x)
        Y.append(y)
        
len(X)
```

```python
with open('org-augmentation.json', 'w') as fopen:
    json.dump({'X': X, 'Y': Y}, fopen)
```
