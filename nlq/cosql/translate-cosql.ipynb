```python
import json

with open('cosql.json') as fopen:
    data = json.load(fopen)
```

```python
translators = translate_sparc.Translate_Concurrent(batch_size = 20, from_lang = 'en', to_lang = 'ms')
```

```python
t = translators.translate_batch(data)
```

```python
with open('cosql.json.translate', 'w') as fopen:
    json.dump(t, fopen)
```

```python
data = []

for i in range(len(t)):
    d = t[i][0].copy()
    d['question_bahasa'] = t[i][1]
    data.append(d)
```

```python
with open('cosql.json.data', 'w') as fopen:
    json.dump(data, fopen)
```
